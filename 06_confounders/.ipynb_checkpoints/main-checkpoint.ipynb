Luego, para tratar de entender cómo las variables actúan sobre los cargos de los clientes se prueba un modelo de regresión considerando la edad, tabaquismo y obesidad. Se obtiene el gráfico con las regresiones lineales sobre cada grupo utilizando modelos lineales.

```r
modelo_1 <- lm(charges ~ age + smoker + obesidad, data=df_seguros)

# Agregar datos del modelo
grid_1 <- modelr::data_grid(df_seguros, age, smoker, obesidad)
# Agregar predicciones del modelo
grid_1 <- modelr::add_predictions(grid_1, modelo_1)

# Gráfico del modelo
ggplot(df_seguros, aes(x=age, y=charges, color=smoker, shape=obesidad)) + geom_point() + 
    geom_line(data=grid_1, mapping=aes(y=pred))
```

##### ***Figura 3.*** "Porcentaje de tabaquismo, obesidad, tabaquismo y obesidad, o sin condiciones preexistentes en las poblaciones de las cuatro regiones estudiadas.

```r
df_seguros <- mutate(df_seguros, TO = (smoker=='yes') & obesidad)
modelo <- lm(charges ~ age +  smoker*obesidad, data=df_seguros)

# Agregar datos del modelo
grid <- modelr::data_grid(df_seguros, age, smoker, obesidad)
# Agregar predicciones del modelo
grid <- modelr::add_predictions(grid, modelo)

# Gráfico del modelo
p <- ggplot(df_seguros, aes(x=age, y=charges, color=smoker, shape=obesidad)) + geom_point() + 
    geom_line(data=grid, mapping=aes(y=pred, color=smoker, linetype=obesidad))
```
