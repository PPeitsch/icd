```r
# Third model
df_model_3 <- reduced_data
model_3 = lm(Close_btc ~ (Close_oro + Close_wti + Close_nasdaq)*Date, data=df_model_3)
df_model_3 <- add_predictions(data=df_model_3, model=model_3) %>% add_residuals(model=model_3)

# Fourth model
df_model_4 <- reduced_data
model_4 = lm(Close_btc ~ (Close_oro + Close_wti + Close_nasdaq + Volume_btc)*Date, data=df_model_4)
df_model_4 <- add_predictions(data=df_model_4, model=model_4) %>% add_residuals(model=model_4)

# Combining the dataframes
df_model_3_4 <- rbind(transform(df_model_3, Model = "Model 3"), transform(df_model_4, Model = "Model 4"))
```

```r
# Third model
m3 <- ggplot(data=df_model_3) + 
        geom_point(aes(x=Date, y=Close_btc)) +
        geom_line(data=df_model_3, aes(x=Date, y=pred), color='DarkGreen', size=1.5) +
        theme(axis.line = element_line(colour = "black", size = 1), text = element_text(size = 30),
              plot.title = element_text(color="Black", size=30, face="bold")) +
        labs(x="Fecha (años)", y="Precio (u$s)", title="Regresion lineal: modelo 3")
m3_name <- "m3.jpeg"
```

```r
# Fourth model
m4 <- ggplot(data = df_model_3_4) +
        geom_point(aes(x = Date, y = Close_btc), color = "DarkGreen", size=2) +
        geom_line(aes(x = Date, y = pred, color = Model), size = 1) +
        theme(axis.line = element_line(colour = "black", size = 1), text = element_text(size = 30),
        plot.title = element_text(color = "Black", size = 30, face = "bold")) +
        labs(x = "Fecha (años)", y = "Precio (usd)", title = "Regresión lineal: modelos 3 y 4")
m4_name <- "m4.jpeg"
```
